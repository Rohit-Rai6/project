```python
import pandas as pd

import requests

from bs4 import BeautifulSoup

```

```python
site=requests.get('https://housing.com/in/buy/searches/AE0P38f9yfbk7p3m2h1f')
print(site.content)
```

```python
data=BeautifulSoup(site.content,'html.parser')
print(data.prettify())
```

```python
productname=data.find_all('div',class_="css-fyni2o")
print(productname)
```

```python
ProjectName=[]
for i in data.find_all('div',class_="_9s1o8l _vhjp4b _gzftgi _l81s5a _3f1mtk _c8sm30 _g3exct _cs1nn1 _frwh2y _2djk66 _19381tsp _1h3d1jbn _pf1rftgi _1lid1y44 _1n00oii0"):
    productname=i.text
    ProjectName.append(productname)
print(ProjectName)
```

```python
Name=[]
for a in data.find_all('h2',class_='_gi182v _1u71grho _7s5wglyw'):
    ratingname=a.text
    Name.append(ratingname)
print(Name)
```

```python
Price=[]
for b in data.find_all('div',class_="_csbfng _c8f6fq _g3gktf _ldyh40 _7l1ulh"):
    pricename=b.text
    Price.append(pricename)
print(Price)
```

```python
EMI=[]
for c in data.find_all('span',class_="_9jtlke _l881bl _gzftgi _7l1994 _c81fwx _g3dlk8"):
    eminame=c.text
    EMI.append(eminame)
print(EMI)
```

```python
print(len(ProjectName))
print(len(Name))
print(len(Price))
```

```python
Housing=pd.DataFrame({'Project Name':ProjectName,'Building Name':Name,'Pricing':Price,})
Housing
```
